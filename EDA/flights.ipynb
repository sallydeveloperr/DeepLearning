```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights.info()
```

```python
flights.T
```

```python
flights.pivot(index='year',columns='month',values='passengers')
```

```python
# 연도별 승객수의 변화 lineplot - sns
import seaborn as sns
import matplotlib.pylab as plt
sns.lineplot(data=flights, x='year', y = 'passengers')
# sns.lineplot(data=flights, x='year', y = 'passengers',errorbar=('ci', 85))
# sns.lineplot(data=flights, x='year', y = 'passengers',errorbar=None) 음영없이 선만 있는 버전
# sns 시각화는 신뢰구간도같이 표시하는 경우
# 데이터가 변동될 수 있는 구간
# 신뢰구간이 넓으면 불확실성이 크다. 데이터 분산이 크거나 샘플수가 작을 경우
# 신뢰구간이 좁으면 안정적이다. 데이터 모여있거나 샘플수가 많은 경우 
# plt.legend(loc='best')
plt.show()
```

```python
# 년도별 월별 승객의 변화 시각화
plt.figure(figsize=(12,5))
sns.lineplot(data=flights,x='year',y='passengers',hue='month')
```
