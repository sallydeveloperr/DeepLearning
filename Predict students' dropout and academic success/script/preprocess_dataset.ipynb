```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# 1️⃣ 데이터 불러오기
df = pd.read_csv("C:\\Users\\sally\\SKNPYWORKSPACE\\Predict students' dropout and academic success\\raw\\dataset.csv")

# 2️⃣ Target 인코딩 (학생 이탈 여부)
df['Target'] = df['Target'].map({'Dropout': 1, 'Enrolled': 0, 'Graduate': 0})

# 3️⃣ 경제지표 제거 (선택)
df = df.drop(columns=['Unemployment rate', 'Inflation rate', 'GDP'])

# 4️⃣ 컬럼 구분
categorical_cols = [
    'Marital status', 'Application mode', 'Application order', 'Course',
    'Daytime/evening attendance', 'Previous qualification', 'Nacionality',
    "Mother's qualification", "Father's qualification", "Mother's occupation",
    "Father's occupation", 'Displaced', 'Educational special needs', 'Debtor',
    'Tuition fees up to date', 'Gender', 'Scholarship holder', 'International'
]

numeric_cols = [
    'Age at enrollment',
    'Curricular units 1st sem (credited)', 'Curricular units 1st sem (enrolled)',
    'Curricular units 1st sem (evaluations)', 'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)', 'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)', 'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)', 'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)', 'Curricular units 2nd sem (without evaluations)'
]

# 5️⃣ 원-핫 인코딩 (sparse_output=True)
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded = encoder.fit_transform(df[categorical_cols])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols))

# 6️⃣ 수치형 변수 스케일링
scaler = StandardScaler()
scaled = scaler.fit_transform(df[numeric_cols])
scaled_df = pd.DataFrame(scaled, columns=[col + "_scaled" for col in numeric_cols])

# 7️⃣ 병합
processed_df = pd.concat([encoded_df, scaled_df, df['Target'].reset_index(drop=True)], axis=1)

# 8️⃣ 저장
processed_df.to_csv("C:\\Users\\sally\\SKNPYWORKSPACE\\student_retention_preprocessed.csv", index=False)

# 9️⃣ 확인
print("✅ 전처리 완료")
print("최종 데이터 크기:", processed_df.shape)
print(processed_df.head())
```
